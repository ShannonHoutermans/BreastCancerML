### Scaling
For this research, we are using the *K-Nearest Neighbour* algorithm to predict the class per observation, based on the given variables. In KNN, the prediction is made by estimating the likelyhood that a data point belongs to one class or the other based on the class of the nearest datapoints.<sup>8</sup>

The distance to the nearest datapoints is calculated using Euclidian distance, which is the absolute distance between two values. In order to ensure that the same unit of measurement is used among features, and that no features weigh more than others, we want to transform the values into the same unit of measurement, which will be Standard Deviation (SD) for this project.

Below, we use a scaler called `StandardScaler` in order to transform the values, after which we can visualise the results in a scatterplot.
